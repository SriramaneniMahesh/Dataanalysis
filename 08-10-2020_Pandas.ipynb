### Data Frame:

```python
# create a dataframe

import pandas as pd
import numpy as np

df=pd.DataFrame()
type(df)
```

```python
# Assign the data using list into DF
l=[10,20,15,25,36]
df=pd.DataFrame(l)
df
```

```python
l=[[10,20],[30,40],[50,60],[70,80]]
df=pd.DataFrame(l)
df
```

```python
l=[[10,20],[30,40],[50,60],[70,80]]
df=pd.DataFrame(l,columns=["first","second"])
df
```

```python
# pass the data using numpy into dataframe
a=np.array(['dharma teja','s kumar','pavan kumar','shankar'])
df=pd.DataFrame(a,columns=["Name"])
df
```

```python
# pass the list of tuples into dataframe

l=[("A",30),("B",40),("c",50),("D",70)]
df=pd.DataFrame(l)
df.columns=["Name","value"]
df
```

```python
# pass the data into df by using dictionary
d={"sno":[101,102,103,104],"sname":["A","B","C","D"]}
df=pd.DataFrame(d)
df
```

```python
# update the column name
df.columns=["Student Number","Student Name"]
df
```

```python
# How to add the new columns
df['Student Marks']=[50,60,70,25]
df
```

```python
# How to add the new row 
new_row={"Student Number":105,"Student Name":"E","Student Marks":70}
df=df.append(new_row,ignore_index=True)
df
```

```python
d={"sno":[101,102,103,104,105],'sname':["A","B","c","D","E"],'smarks':[50,60,70,80,30]}
df=pd.DataFrame(d,index=[1,2,3,4,5])
df
```

### 0-25 => fail
### 25-60 => second
### 60 above=>first

```python
l=[]
for i in df['smarks']:
    if i>0 and i<=25:
        l.append("Fail")
    elif i>25 and i<60:
        l.append("Second")
    else:
        l.append("First")
df['student result']=l
df
```

```python
df['student result']=='Fail'
```

```python
df[df['student result']=='Fail']
```

```python
df[df['smarks']>50]
```

```python
df[(df['smarks']<22)&(df['student result']=='Fail')]
```

```python
# Delete the columns
df=df.drop('student result',axis=1)
df
```
